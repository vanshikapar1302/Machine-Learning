```python
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}

    for user in data:
        raw_rating = user.get("rating", None)

        if raw_rating is None:
            print("Rating missing")
            continue

        raw_rating = raw_rating.strip().lower()
        if raw_rating in text_to_num:
            clean_rating = text_to_num[raw_rating]
        else:
            clean_rating = float(raw_rating)

        print(clean_rating)
```

```python
clean_data(data)
```
