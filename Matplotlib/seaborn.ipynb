```python
tips=sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="smoker"
)
    
    
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line"
)
    
    
```

```python
flights=sns.load_dataset("flights")
#flights.head()
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

```python
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
penguins=sns.load_dataset("penguins")
#penguins.head()
sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)
```

```python
flights
```

```python
pivot_data=flights.pivot(index="month",columns="year",values="passengers")
print(pivot_data)
```

```python
sns.heatmap(
    pivot_data,
    cmap="coolwarm"
)
```

```python
sns.heatmap(
    pivot_data,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)
plt.title("Passengers count in different years")
plt.tight_layout()
```

```python
fig, ax = plt.subplots()

sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    ax=ax,
    errorbar=None #confidence interval(ci) approx wlla ya most probable data hide krega
)
ax.set_title("Average Total Bill by Day and Gender")
ax.set_xlabel("Day")
ax.set_ylabel("Total Bill ($)")
fig.tight_layout()
```

```python
tips = sns.load_dataset("tips")

plt.figure()
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.title("Total Bill Distribution by Day and Sex")
plt.show()
```
