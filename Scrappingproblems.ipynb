```python
from bs4  import BeautifulSoup
```

```python

page_count = 1
while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)

    soup = BeautifulSoup(res.text, "html.parser")
    quotes = soup.select("div.quote")

    # stop when no quotes found (last page)
    if not quotes:
        print("No valid pages anymore.")
        break

    # save page html
    with open(f"scrapped_data/quotes_{page_count}.html", "w", encoding="utf-8") as f:
        f.write(res.text)

    print(f"Downloaded data from page {page_count}")

    page_count = page_count + 1
```

```python
page_count = 1

with open(f"scrapped_data/quotes_{page_count}.html", "r", encoding="utf-8") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "html.parser")
```

```python
all_quotes = soup.select("div.quote")
life_quotes = []

for q in all_quotes:
    all_tags = []

    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())

    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append([text, author])

print(life_quotes)
```

```python
from bs4 import BeautifulSoup
import os

folder = "scrapped_data"
life_quotes = []

for file_name in os.listdir(folder):
    if file_name.endswith(".html"):

        with open(os.path.join(folder, file_name), "r", encoding="utf-8") as f:
            html = f.read()

        soup = BeautifulSoup(html, "html.parser")
        all_quotes = soup.select("div.quote")

        for q in all_quotes:
            tags = [tag.get_text() for tag in q.select(".tags .tag")]

            if "life" in tags:
                text = q.select_one("span.text").get_text()
                author = q.select_one("small.author").get_text()
                life_quotes.append([text, author])

print("Total life quotes:", len(life_quotes))
for q in life_quotes:
    print(q[0], "-", q[1])
```
