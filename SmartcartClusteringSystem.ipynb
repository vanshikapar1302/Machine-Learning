```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("smartcart_customers.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

# Data Preprocessing

```python
#handle missing values
df["Income"]=df["Income"].fillna(df["Income"].median())
```

```python
df.isnull().sum()
```

# Feature engineering

```python

df.columns
```

```python


# Age
df["Age"]=2026-df["Year_Birth"]
```

```python
df.head()
```

```python
#Customer Joining date
df["Dt_Customer"]=pd.to_datetime(df["Dt_Customer"],dayfirst=True)
refernce_date=df["Dt_Customer"].max()
df["Customer_tenure_days"]=(refernce_date-df["Dt_Customer"]).dt.days
```

```python
#Spending
df["Total_spending"]=df["MntWines"]+df["MntGoldProds"]+df["MntFishProducts"]+df["MntFruits"]+df["MntSweetProducts"]+df["MntMeatProducts"]
```

```python
# Total Child
df["Total_children"]=df["Kidhome"]+df["Teenhome"]
```

```python
df["Education"] = df["Education"].replace({
    "Basic": "Undergraduate",
    "2n Cycle": "Undergraduate",
    "Graduation": "Graduate",
    "Master": "Postgraduate",
    "PhD": "Postgraduate"
})
```

```python
df["Education"].value_counts()
```

```python
df["Marital_Status"].value_counts()
```

```python
# Marital Status cleaning

df["Living_With"] = df["Marital_Status"].replace({
    "Married": "Partner",
    "Together": "Partner",
    "Single": "Alone",
    "Divorced": "Alone",
    "Widow": "Alone",
    "Absurd": "Alone",
    "YOLO": "Alone"
})
```

```python
df["Marital_Status"].value_counts()
```

# Drop columns

```python
cols = ["ID","Year_Birth","Marital_Status","Kidhome","Teenhome","Dt_Customer"]

spending_cols = ["MntWines","MntFruits","MntMeatProducts",
                 "MntFishProducts","MntSweetProducts","MntGoldProds"]

cols_to_drop = cols + spending_cols

df_cleaned = df.drop(columns=cols_to_drop)
```

```python
df_cleaned.shape
```

```python
df_cleaned.head()
```

# Outliers

```python
cols=["Income","Recency","Response","Age","Total_spending","Total_children"]
sns.pairplot(df_cleaned[cols])
```

```python
 # Remove outliers

print("Data size with outliers:", len(df_cleaned))

df_cleaned = df_cleaned[(df_cleaned["Age"] < 90)]
df_cleaned = df_cleaned[(df_cleaned["Income"] < 600000)]

print("Data size without outliers:", len(df_cleaned))
```

# Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
cat_cols=["Education","Living_With"]
enc_cols=ohe.fit_transform(df_cleaned[cat_cols])
```

```python
enc_df=pd.DataFrame(enc_cols.toarray(),columns=ohe.get_feature_names_out(cat_cols),index=df_cleaned.index)
```

```python
enc_df.head()
```

```python
df_encoded=pd.concat([df_cleaned.drop(columns=cat_cols),enc_df],axis=1)
```

```python
df_encoded.shape
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
X=df_encoded
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

# Visualization

```python
X_scaled.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)

X_pca=pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection="3d")
plt.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2])

ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
ax.set_title("3D projection")
```

# Analyze K value

# 1. Elbow method

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit_predict(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
knee=KneeLocator(range(1,11),wcss,curve="convex",direction="decreasing")
optimal_k=knee.elbow
```

```python
print("Optimal k:",optimal_k)
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel("k")
plt.ylabel("WCSS")
```

# 2.Silhouette Score

```python
from sklearn.metrics import silhouette_score
scores=[]
for k in range(2,11):
      kmeans=KMeans(n_clusters=k,random_state=42)
      labels=kmeans.fit_predict(X_pca)
      score=silhouette_score(X_pca,labels)
      scores.append(score)
    
```

```python
plt.plot(range(2,11),scores,marker='o')
plt.xlabel("K")
plt.ylabel=("Silhouette score")
```

# Clustering

```python
# K_means

kmeans = KMeans(n_clusters=4, random_state=42)
labels_kmeans = kmeans.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels_kmeans)
```

```python
# Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering
agg_clf = AgglomerativeClustering(n_clusters=4, linkage="ward")
labels_agg = agg_clf.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels_agg)
```
