```python
 from bs4   import BeautifulSoup
with open ("Scrapped_data/data1.html","r", encoding="utf-8") as f:
    html_content=f.read()

soup=BeautifulSoup(html_content,"lxml")
soup.find("h1")
```

```python
all_h3=soup.find_all("h3")
for h3 in all_h3:
    name=h3.get_text(strip=True)
    print(name)
```

```python
for h3 in all_h3:
    parent = h3.find_parent("div")
    population = parent.select_one("span.country-population")

    if population:   # list is NOT empty
        print(population.get_text())
```

```python
data = []

for h3 in all_h3:
    name = h3.get_text(strip=True)

    parent = h3.find_parent("div")
    pop_tag = parent.select_one("span.population")

    if pop_tag:
        population = pop_tag.get_text(strip=True)
    else:
        population = "N/A"

    data.append([name, population])

print(len(data))
print(data[:10])
```
